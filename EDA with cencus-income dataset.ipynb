# Number of people with income ratio.

```python
plt.figure(figsize=(10,10))
plt.title("People income >50k and <=50k")
sns.countplot(x=a['Listing_of_attributes'])
a['Listing_of_attributes'].value_counts(ascending=False).values

```

# Univariate analysis for each numerical columns.

```python
plt.figure(figsize=(15,15))
plt.title("Univariate analysis for each numerical columns")
for i in range (0,len(numerical_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=a[numerical_features[i]],shade=True,color='b')
    plt.xlabel(numerical_features[i])
    
```

# Histograms-plot based on the all-categories columns.

```python
plt.figure(figsize=(40,40))
plt.title("Univariate analysis for each numerical columns")
for i in range (0,len(categorical_features)):
    plt.subplot(5,3,i+1)
    sns.countplot(x=a[categorical_features[i]],color='b')
    plt.xlabel(categorical_features[i])
```
